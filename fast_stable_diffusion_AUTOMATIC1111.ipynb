```python
#@markdown # Bind mount google drive to local storage
#@markdown This makes the relevant directories and files from your google drive appear in your local storage. The following directories are mounted:\
#@markdown <small>Paths are not copyable because google colab [jupyter notebook wants them to be clickable](https://github.com/ipython/ipython/pull/2464)), even in code blocks and other methods of preformatting.</small>
#
#@markdown |Directory/file|contents|Location (in your google drive)|
#@markdown |---|---|---|
#@markdown |extensions/|webui extensions|`/ sd / stable-diffusion-webui / extensions`|
#@markdown |models/|This has subdirectories for Loras, VAE, diffusion models, upscalers, and so on.|`/ sd / stable-diffusion-webui / models/`|
#@markdown |embeddings/|textural inversions. I don't know why these aren't in the models directory.|`/ sd / stable-diffusion-webui / embeddings`|
#@markdown |outputs/|images that you generate|`/ outputs`|
#@markdown |config.json|I believe this file changes what options are available in the UI, as well as...idk...<br/>When I changed my quicksettings and other options, this file changed.|`/ sd / stable-diffusion-webui / config.json`|
#@markdown |ui-config.json|I think this file lets you change defaults? not sure|`/ sd / stable-diffusion-webui / ui-config.json`|
# first unmount any that are already mounted
with capture.capture_output() as cap:
  !for a in extensions models embeddings outputs config.json ui-config.json; do umount /content/gdrive/MyDrive/sd/stable-diffusion-webui/$a; done
  # rename any that weren't bind-mounted (don't delete anything, I learned this the hard way...delete from inside drive if I do something weird.)
  !for a in extensions models embeddings config.json ui-config.json; do mv /content/gdrive/MyDrive/sd/stable-diffusion-webui/$a /content/gdrive/MyDrive/sd/stable-diffusion-webui/$a.bak; done
  # make new empty directories
  !for a in extensions models embeddings outputs; do mkdir -p /content/gdrive/MyDrive/sd/stable-diffusion-webui/$a; done > /dev/null
  # bind-mount drive directories to them
  !for a in extensions models embeddings; do mount --bind $(realpath /content/gdrive2/MyDrive/sd/stable-diffusion-webui/$a) /content/gdrive/MyDrive/sd/stable-diffusion-webui/$a; done
  # bind-mount output directory to unshared drive folder (also make sure it exists)
  !mkdir -p /content/gdrive2/MyDrive/outputs; mount --bind $(realpath /content/gdrive2/MyDrive/outputs) /content/gdrive/MyDrive/sd/stable-diffusion-webui/outputs;
  # bind-mount configs (http://www.mardy.it/blog/2018/10/how-to-bind-mount-single-file.html)
  # but only if they already exist...if they don't exist, I think they get autogenerated, in which case you'll want to copy them over.
  !for a in config.json ui-config.json params.txt; do [ -e "/content/gdrive2/MyDrive/sd/stable-diffusion-webui/$a" ] && touch "/content/gdrive/MyDrive/sd/stable-diffusion-webui/$a" && mount -o rw,bind "/content/gdrive2/MyDrive/sd/stable-diffusion-webui/$a" "/content/gdrive/MyDrive/sd/stable-diffusion-webui/$a"; done
  # lastly, delete that annoyingly distracting sample data directory
  !rm -rf /content/sample_data
```
